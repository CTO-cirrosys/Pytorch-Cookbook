# Transform을 이용한 Augmentation
- Augmentation이란 데이터를 다양하게 만들어주는 작업
- Pytorch에서 제공하는 Transform를 이용해 다양한 Augmentation 시도

```python
from PIL import Image
import torch
import torchvision.transforms as transform
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = '../data/example.jpg'

img = Image.open(path)

plt.figure(figsize = (10,8))
plt.axis('off')
plt.imshow(img)
```

```python
# 중심 자르기
img1 = transform.CenterCrop(size = (100,100))(img)# size에 맞게 중심 자르기
img1
```

```python
# 밝기, 명암 바꿔주기
img2 = transform.ColorJitter(brightness=1, contrast=0, saturation=0, hue=0)(img) # 특정 수치 범위내에서 랜덤으로 바꿔준다
img2
```

```python
img2.save(path + 'b.jpg')
```

```python
# Image을 5개로 나누기
fig = plt.figure(figsize = (20,30))
for i,im in enumerate(transform.FiveCrop(size = (50,50))(img)): # 설정한 크기로 5개의 사진으로 나눔
    plt.subplot(1,5,i+1)
    plt.imshow(im)
    plt.axis('off')
    
```

```python
fig.savefig(path + 'img2.jpg')
```

```python
# Gray 스케일로 바꾸기
img3 = transform.Grayscale(num_output_channels=1)(img)
img3
```

```python
# Image을 5개로 나누기
fig = plt.figure(figsize = (20,30))
mode = ['constant', 'edge', 'reflect']

for i, m in enumerate(mode): # 설정한 크기로 5개의 사진으로 나눔
    plt.subplot(1,3,i+1)
    im = transform.Pad(padding=(100,100), fill=(255,0,255), padding_mode=m)(img)
    plt.imshow(im)
    plt.title(m)
    plt.axis('off')
```

```python
# 이미지 회전
img5 = transform.RandomAffine(degrees=90, fillcolor=255)(img) # fiilcolor는 PIL >= 5.0일때만 작동
img5
```

```python
# 여러 기능을 랜덤하게 적용하기(모두 적용이 된다.)
transforms = [
    transform.ColorJitter(brightness=1),
    transform.RandomAffine(degrees=180),
    transform.CenterCrop(size=(50,50))
]

transform.RandomApply(transforms, p=0.5)(img)
```

```python
# 여러 기능을 랜덤하게 적용(여러개의 옵션 중 선택)
transform.RandomChoice(transforms)(img)
```

```python
# 랜덤으로 자르기
img6 = transform.RandomCrop(size=(100,120))(img)
img6
```

```python
img6.save(path + 'img6.jpg')
```

```python
# 랜덤하게 grayscale
transform.RandomGrayscale(p=0.5)(img)
```

```python
# 랜덤하게 뒤집기
img8 = transform.RandomHorizontalFlip(p=0.9)(img) # 적용되면 사진이 살짝 어색.....
img8
```

```python
img8.save(path + 'img8.jpg')
```

```python
# image 사이즈 변경(보간법에 따라 다름, 숫자로 구별)
print(transform.Resize((100,100), interpolation=1))
print(transform.Resize((100,100), interpolation=2))
print(transform.Resize((100,100), interpolation=3))
print(transform.Resize((100,100), interpolation=4))
print(transform.Resize((100,100), interpolation=5))

transform.Resize((100,100), interpolation=2)(img)
```

```python
# 이미지에서 랜덤으로 부위 삭제
tensor = transform.ToTensor()(img)
example = transform.RandomErasing(p=0.8, scale=(0.02, 0.33), ratio=(0.3, 3.3), value=0)(tensor)
plt.imshow(example.permute(1,2,0))
```

```python
# Tensor 형식으로 바꿔주기
transform.ToTensor()(img) # Totensor는 자동으로 0~1사이 값으로 바꿔줌
```

```python
# Normalize도 가능
transform.Normalize(mean=(0,0,0), std=(2,2,2))(tensor)
```
