```python
import matplotlib.pyplot as plt
import numpy as np
```

# 계단함수

```python
def f(x):
    y = x > 0
    return y.astype(np.int)

x = np.arange(-5, 5.0, 0.1)
y = f(x)
plt.plot(x, y)
plt.savefig('stair.jpg')
```

# 시그모이드 함수

```python
def f(x):
    y = 1.0/(1.0 + np.exp(-x))
    return y

x = np.arange(-5, 5.0, 0.1)
y = f(x)

plt.plot(x,y)
plt.savefig('sig.jpg')
```

# Tanh

```python
def f(x):
    y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
    return y

x = np.arange(-5, 5.0, 0.1)
y = f(x)

plt.plot(x,y)
plt.savefig('tanh.jpg')
```
